```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "Student_ERP": [6605563, 6605516, 6605342, 6605523, 6605687, 6605761],
    "Name": ["Ashutosh", "Rahul", "Priya", "Neha", "Aman", "Karan"],
    "Marks": [78, 85, 92, 60, 55, 68]
}

df = pd.DataFrame(data)
df
```

```python
mean_marks = df["Marks"].mean()
mean_marks
```

```python
std_marks = df["Marks"].std()
std_marks
```

```python
df["Z_Score"] = (df["Marks"] - mean_marks) / std_marks
df
```

```python
plt.hist(df["Marks"])
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

```python
import gradio as gr

def analyze_exam():
    mean_marks = df["Marks"].mean()
    std_marks = df["Marks"].std()

    df["Z_Score"] = (df["Marks"] - mean_marks) / std_marks

    plt.figure()
    plt.hist(df["Marks"])
    plt.title("Marks Distribution")
    plt.xlabel("Marks")
    plt.ylabel("Students")
    plt.show()

    summary = f"""
    Mean Marks: {mean_marks:.2f}
    Standard Deviation: {std_marks:.2f}
    """

    return df, summary

interface = gr.Interface(
    fn=analyze_exam,
    inputs=[],
    outputs=[
        gr.Dataframe(label="Exam Result Analysis"),
        gr.Textbox(label="Statistical Summary")
    ],
    title="Exam Result Statistical Report Generator",
    description="AI/ML based statistical analysis of exam results"
)

interface.launch()
```
